```python
import matplotlib.pyplot as plt
import pandas as pd
import scipy.stats as st
import numpy as np

mouse_metadata_path = "data/Mouse_metadata.csv"
study_results_path = "data/Study_results.csv"

mouse_metadata = pd.read_csv(mouse_metadata_path)
study_results = pd.read_csv(study_results_path)


pyma_df = pd.merge(study_results, mouse_metadata, on = ['Mouse ID'], how = 'right')

pyma_df.head()
```

```python
unique_mice_df = pyma_df['Mouse ID'].nunique()
print(unique_mice_df)
```

```python
duplicate_mouse_ids = pyma_df[pyma_df.duplicated(subset=['Mouse ID', 'Timepoint'])]['Mouse ID'].unique()
duplicate_mouse_ids
```

```python
duplicate_mice = pyma_df[pyma_df.duplicated(subset=['Mouse ID', 'Timepoint'])]
duplicate_mice
```

```python
cleaned_df = pyma_df[pyma_df['Mouse ID'] != 'g989']
cleaned_df.head()
```

```python
c_unique_mice_df = cleaned_df['Mouse ID'].nunique()
print(c_unique_mice_df)
```

```python
# Generate a summary statistics table of mean, median, variance, standard deviation, and SEM of the tumor volume for each regimen

# Use groupby and summary statistical methods to calculate the following properties of each drug regimen: 
# mean, median, variance, standard deviation, and SEM of the tumor volume. 
summary_df = pyma_df.groupby('Drug Regimen')
summary_stats = summary_df['Tumor Volume (mm3)'].agg(['mean', 'median', 'var', 'std', 'sem'])

# Rename the columns
summary_stats = summary_stats.rename(columns={
    'mean': 'Mean Tumor Volume',
    'median': 'Median Tumor Volume',
    'var': 'Tumor Volume Variance',
    'std': 'Tumor Volume Std. Dev.',
    'sem': 'Tumor Volume Std. Err.'
})
# Assemble the resulting series into a single summary DataFrame.
summary_stats
```

```python
#Grouping the DataFrame by Drug Regimen
regimen_observation_counts = cleaned_df['Drug Regimen'].value_counts()

# Set x axis and tick
x_axis = np.arange(len(regimen_observation_counts))
tick_locations = [value for value in x_axis]

# List to indicate x labels and set figure size
plt.figure(figsize=(10,6))
plt.bar(x_axis, regimen_observation_counts, color='b', alpha=0.75, align="center")
plt.xticks(tick_locations, regimen_observation_counts.index, rotation="vertical")

#Label names
plt.xlabel("Drug Regimen")
plt.ylabel("# of Obvserved Mouse Timepoints")

#Graph cleanup
plt.xlim(-0.75, len(x_axis)-0.25)
plt.ylim(0, max(regimen_observation_counts) + 10)

# Generate a bar plot showing the total number of rows (Mouse ID/Timepoints) for each drug regimen using Pandas.
plt.show()
```

```python
# Calculate the final tumor volume of each mouse across four of the treatment regimens:  
# Capomulin, Ramicane, Infubinol, and Ceftamin
pd.options.mode.chained_assignment = None
cleaned_df.rename(columns={'Timepoint': 'Last Timepoint'}, inplace=True)
# Start by getting the last (greatest) timepoint for each mouse
last_timepoints = cleaned_df.groupby('Mouse ID')['Last Timepoint'].max()
last_timepoints_df = pd.DataFrame(last_timepoints).reset_index()
last_timepoints_df.head()
```

```python
merged_data = pd.merge(cleaned_df, last_timepoints_df, on=['Mouse ID', 'Last Timepoint'], how='inner').copy()
treatment_regimens = ['Capomulin', 'Ramicane', 'Infubinol', 'Ceftamin']
final_tumor_volume_df = merged_data[merged_data['Drug Regimen'].isin(treatment_regimens)]
final_tumor_volume_df
```

```python
# Put treatments into a list for for loop (and later for plot labels)
treatment_regimens = ['Capomulin', 'Ramicane', 'Infubinol', 'Ceftamin']

# Create empty list to fill with tumor vol data (for plotting)
tumor_volume_data = []

# Calculate the IQR and quantitatively determine if there are any potential outliers. 
for regimen in treatment_regimens:
    regimen_data = final_tumor_volume_df['Drug Regimen']
    # Locate the rows which contain mice on each drug and get the tumor volumes
    tumor_volumes = final_tumor_volume_df['Tumor Volume (mm3)']
    
    # add subset 
    quartiles = tumor_volumes.quantile([0.25, 0.75])
    lower_q = quartiles[0.25]
    upper_q = quartiles[0.75]
    iqr = upper_q - lower_q
    
    # Determine outliers using upper and lower bounds
    lower_bound = lower_q - 1.5 * iqr
    upper_bound = upper_q + 1.5 * iqr
    outliers = tumor_volumes[(tumor_volumes < lower_bound) | (tumor_volumes > upper_bound)]
    

    # Print outliers
    if not outliers.empty:
        print(f"Potential outliers for {regimen}: {outliers}")
    else:
        print(f"No potential outliers for {regimen}")
```

```python
# Generate a box plot that shows the distrubution of the tumor volume for each treatment group.
grouped_data = final_tumor_volume_df.groupby('Drug Regimen')['Tumor Volume (mm3)'].apply(list)
plt.figure(figsize=(10,5))
plt.boxplot(grouped_data.values, labels=grouped_data.index)
plt.title('Final Tumor Volume by Drug Regimen')
plt.xlabel('Drug Regimen')
plt.ylabel('Tumor Volume (mm3)')
plt.show
```

```python
mouse_df = cap_df[cap_df['Mouse ID'] == 's185']
mouse_df
```

```python
# Generate a line plot of tumor volume vs. time point for a single mouse treated with Capomulin
x_axis = mouse_df['Last Timepoint']
tumor = mouse_df['Tumor Volume (mm3)']
plt.plot(x_axis, tumor)
plt.xlabel('Timepoint (Days)')
plt.ylabel('Tumor Volume (mm3)')
plt.title('Capomulin treatment for mouse (s185)')
plt.show
```

```python
average_tumor_volume = cap_df.groupby('Mouse ID')['Tumor Volume (mm3)'].mean()
mouse_weight = cap_df.groupby('Mouse ID')['Weight (g)'].mean()
```

```python
# Generate a scatter plot of mouse weight vs. the average observed tumor volume for the entire Capomulin regimen
plt.scatter(mouse_weight, average_tumor_volume, marker="o", facecolors="blue", edgecolors="blue",
            s=mouse_weight, alpha=0.9)
plt.ylabel('Average Tumor Volume (mm3)')
plt.xlabel('Weight (g)')
plt.title('Mouse Weight vs Tumor Volume for Capomulin Mice')
plt.show()
```

```python
# Calculate the correlation coefficient
correlation_coefficient, _ = st.pearsonr(mouse_weight, average_tumor_volume)
print(f"Correlation Coefficient: {correlation_coefficient}")
```

```python
plt.scatter(mouse_weight, average_tumor_volume, marker="o", facecolors="blue", edgecolors="blue",
            s=mouse_weight, alpha=0.9)

slope, intercept, r_value, p_value, std_err = st.linregress(mouse_weight, average_tumor_volume)
predicted_tumor_volume = slope * mouse_weight + intercept

# Plot the regression line
plt.plot(mouse_weight, predicted_tumor_volume, color='red', linewidth=2)
plt.ylabel('Average Tumor Volume (mm3)')
plt.xlabel('Weight (g)')
plt.title('Mouse Weight vs Tumor Volume for Capomulin Mice')
plt.show()
```
